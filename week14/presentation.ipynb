```python
import plotly.express as px

# GapminderデータをCSVファイルから読み込む
csv_file = 'data/newmaster2.csv'  # ファイル名を適切に設定してください
data = pd.read_csv(csv_file)

# アニメーションを作成
fig = px.choropleth(data_frame=data,
                    locations='ISO3',
                    locationmode='ISO-3',
                    color='pop',
                    hover_name='Country name_x',
                    animation_frame='year',
                    projection='natural earth',
                    color_continuous_scale='Viridis',  # カラーマップをViridisに変更,
                    range_color=[0,150000000])

# タイトルを設定
fig.update_layout(title_text='World Population - Gapminder Data',
                  title_font_size=24)

# アニメーションを表示
fig.show()
```

```python
# アニメーションを作成
fig2 = px.choropleth(data_frame=data,
                    locations='ISO3',
                    locationmode='ISO-3',
                    color='life span',
                    hover_name='Country name_x',
                    animation_frame='year',
                    projection='natural earth',
                    color_continuous_scale='Viridis',  # カラーマップをViridisに変更,
                    range_color=[40,80])

# タイトルを設定
fig2.update_layout(title_text='World lifespan - Gapminder Data',
                  title_font_size=24)

# アニメーションを表示
fig2.show()
```
