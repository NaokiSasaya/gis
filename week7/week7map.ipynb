#　埼玉県の人口マップ

```python
import geopandas as gpd
import pandas as pd
```

```python
saitama_ku=gpd.read_file('N03-23_11_230101.geojson')
```

```python
saitama_ku
```

```python
saitama_ku.info()
```

```python
saitama_ku.plot()
```

```python
saitama_pop = pd.read_csv('saitama.csv')
```

```python
saitama_pop['KEY_CODE'] = saitama_pop['KEY_CODE'].astype(int)
```

```python
saitama_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
saitama_ku['KEY_CODE'] = saitama_ku['KEY_CODE'].astype(int)
```

```python
saitama_ku_pop = saitama_ku.merge(saitama_pop,on='KEY_CODE')
```

```python
saitama_ku_pop['人口総数'] = saitama_ku_pop['人口総数'].astype(int)
```

```python
saitama_ku_pop.plot(column='人口総数',legend=True)
```

今回の地図で人口と地図を組み合わせて作ったことで、今後の地図に対する期待が膨らみました！今後他のものにも生かせるように練習していきたいと思います。
