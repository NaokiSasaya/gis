```python
import pandas as pd

# Load the CSV data into DataFrames
df_master = pd.read_csv('data/')
df_additional = pd.read_csv('additional_data.csv')
```

```python
# Sort the main DataFrame by year and country
df_master.sort_values(['year', 'country'], inplace=True)
```

```python
# Merge the data on 'year' and 'country' columns, keeping all rows
df_merged = pd.merge(df_master, df_additional, on=['year', 'country'], how='left')
```

```python
# Forward fill missing values with data from the previous year
df_merged.fillna(method='ffill', inplace=True)
```

```python
# Display the merged DataFrame
print(df_merged)
```
