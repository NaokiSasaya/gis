```python
import plotly.express as px

# Gapminderãƒ‡ãƒ¼ã‚¿ã‚’CSVãƒ•ã‚¡ã‚¤ãƒ«ã‹ã‚‰èª­ã¿è¾¼ã‚€
csv_file = 'data/newmaster2.csv'  # ãƒ•ã‚¡ã‚¤ãƒ«åã‚’é©åˆ‡ã«è¨­å®šã—ã¦ãã ã•ã„
data = pd.read_csv(csv_file)

# ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ã‚’ä½œæˆ
fig = px.choropleth(data_frame=data,
                    locations='ISO3',
                    locationmode='ISO-3',
                    color='pop',
                    hover_name='Country name_x',
                    animation_frame='year',
                    projection='natural earth',
                    color_continuous_scale='Viridis',  # ã‚«ãƒ©ãƒ¼ãƒžãƒƒãƒ—ã‚’Viridisã«å¤‰æ›´,
                    range_color=[0,150000000])

# ã‚¿ã‚¤ãƒˆãƒ«ã‚’è¨­å®š
fig.update_layout(title_text='World Population - Gapminder Data',
                  title_font_size=24)

# ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ã‚’è¡¨ç¤º
fig.show()
```
